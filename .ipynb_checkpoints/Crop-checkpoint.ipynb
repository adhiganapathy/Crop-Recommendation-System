```python


target = data['label'].values.reshape(-1,1)
features = data.drop('label',axis=1)
print(features.shape)

print(target.shape)
```

```python
from sklearn.preprocessing import LabelEncoder
Label = LabelEncoder()
target_df = Label.fit_transform(target)
print(target_df.shape)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split, KFold,cross_val_score
from sklearn.metrics import classification_report
```

```python
x_train,x_test,y_train,y_test=train_test_split(features,target_df,test_size=0.2,random_state=42)
```

### Bagging Classifier

```python
bagging = BaggingClassifier(estimator=DecisionTreeClassifier(),n_estimators=100,max_samples=0.8,random_state=True)
bagging.fit(x_train,y_train.ravel())
ypred = bagging.predict(x_test)
accuracy_DT = accuracy_score(ypred,y_test)
print("Accuracy score:",accuracy_DT)
KFold=10
score = cross_val_score(bagging,features,target_df,cv=KFold)
print("Cross validation score:",score.mean())
```

```python
print("classification report\n",classification_report(ypred,y_test))
```

### RandomForest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
RF = RandomForestClassifier(n_estimators=100,random_state=42)
RF.fit(x_train,y_train)
ypred =RF.predict(x_test)
accuracy_RF = accuracy_score(ypred,y_test)
print("Accuracy score:",accuracy_RF)
KFold=10
score = cross_val_score(RF,features,target_df,cv=KFold)
print("Cross validation score:",score.mean())
```

```python
print("classification report\n",classification_report(ypred,y_test))
```
